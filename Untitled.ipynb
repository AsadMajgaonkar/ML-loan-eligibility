```python
# reading csv
import pandas as pd
df=pd.read_csv('Loan_Train.csv')
df
```

```python
# checking for null values
df.isnull().sum()
```

```python
# removing null values
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
# plotting graph
import seaborn as sns
sns.countplot(x='Gender', hue='Loan_Status', data=df)
```

```python
# replacing values
df.replace({"Loan_Status":{'N':0,'Y':1}},inplace=True)
df.replace({"Gender":{'Female':0,'Male':1},"Education":{'Graduate':0,'Not Graduate':1},"Married":{'No':0,'Yes':1}},inplace=True)
df.replace({"Self_Employed":{'No':0,'Yes':1},"Property_Area":{'Rural':0,'Semiurban':1,'Urban':2}},inplace=True)
df = df.replace(to_replace='3+', value=4)
```

```python
# dividing into input and output dataframes
X= df.drop(columns=['Loan_ID', 'Loan_Status'])
Y= df['Loan_Status']
```

```python
X
```

```python
Y
```

```python
# Decision Tree Algorithm
from sklearn.tree import DecisionTreeClassifier

model= DecisionTreeClassifier()
model.fit(X,Y)
predictions = model.predict([X.values[1]])
predictions
```

```python
# testing and accuracy
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
X_train, X_test, Y_train, Y_test= train_test_split(X,Y,test_size=0.2)
model.fit(X_train, Y_train)
predictions = model.predict(X_test)
score = accuracy_score(predictions, Y_test)
score
```
